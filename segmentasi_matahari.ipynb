```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#untuk membaca lokasi gambar yang diupload
matahari=cv2.imread('matahari.jpg')
#untuk mengconvert BGR ke RGB
matahari=cv2.cvtColor(matahari,cv2.COLOR_BGR2RGB)
#untuk mengconvert RGK ke HSV
hsv_matahari=cv2.cvtColor(matahari, cv2.COLOR_RGB2HSV)
```

```python
#deklarasi rgb batas bawah dengan warna orange muda
orange_muda =(1,190,200)
#deklarasi rgb batas bawah dengan warna orange tua
orange_tua =(18, 255, 255)
```

```python
#Untuk melakukan tresholding
mask=cv2.inRange(hsv_matahari,orange_muda,orange_tua)
#untuk impose gambar asli dengan mask
result=cv2.bitwise_and(matahari,matahari, mask=mask)
```

```python
#plotiing
plt.subplot(1,2,1)
plt.imshow(matahari)
plt.subplot(1,2,2)
plt.imshow(result)
plt.show()
```
